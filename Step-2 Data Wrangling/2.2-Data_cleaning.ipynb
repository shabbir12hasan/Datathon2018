### Taking output of 2.1 file and working ahead on it

Task: Removing stops with all null values

```python
import pandas as pd
import datetime
```

```python
agg_df = pd.read_csv("cleaning/aggregated_stop_data_weekname.csv", delimiter=",")
```

```python
agg_df.head(2)
```

```python
agg_df.shape #Checking shape
```

```python
#removing few columns related to stopID
columns = ["StopType", "SuburbName", "RegionName","GPSLat","GPSLong"]
agg_df = agg_df.drop(columns, 1)
```

```python
agg_df.head(2)
```

```python
# changing axis of the week and stop id, transposing
agg_df.set_index('StopID',inplace=True)
agg_df = agg_df.transpose()
```

```python
agg_df.head()
```

```python
# removing week data from index
agg_df = agg_df.reset_index()
```

```python
#renaming column
agg_df =agg_df.rename(columns = {'StopID':'Index'})
agg_df =agg_df.rename(columns = {'index':'Date'})
```

```python
agg_df.head(2)
```

```python
#converting it to string and then removing week from column
agg_df["Date"] = agg_df["Date"].astype(str).str[5:] 
```

```python
# converting date from week name into date format, 
# taking first day of the week as date

agg_df['Date'] = agg_df['Date'].apply(lambda x: datetime.datetime.strptime(x + '-0', "%Y%W-%w"))
```

```python
agg_df.head(2)
```

```python
#Checking null values
agg_df.isnull().sum().sort_values(ascending=False)
```

```python
# removing column with all null values
df = agg_df.dropna(axis=1,how='all')
```

```python
#Checking null values
null_recs = df.isnull().sum().sort_values(ascending=False)
```

```python
#null_recs
for i in null_recs:
    if i>100:
        print(i)
    #break
```

### Melting data, reducing columns

```python
df = pd.melt(df, id_vars=['Date'])
```

```python
df.columns=['Date','StopID','Counts']
```

```python
df.head()
```

```python
df.shape
```

```python
# changing na to 0
df['Counts'].fillna(0, inplace=True)
```

### Writing data

```python
df.to_csv("cleaning/agg_data_melted.csv", sep=',', index=False)
```
